```kotlin
%use dataframe
```

```kotlin
val genresV2 = DataFrame.readCSV("../data/genres_v2.csv", delimiter = ',')
genresV2
```

```kotlin
genresV2.select { genre }.distinct()
```

```kotlin
val aggregated = genresV2.groupBy { genre }.count()
aggregated
```

```kotlin
aggregated.sortByDesc { count }.plot { 
    bars { 
        x(genre)
        y(count)
    }
}
```
