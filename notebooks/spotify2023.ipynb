```kotlin
%use dataframe
%use kandy
```

```kotlin
val songs = DataFrame.readCSV("../data/spotify-2023-orig.csv", delimiter = ',')
    .rename("artist(s)_name")
    .into("artist")
songs
```

```kotlin
songs.describe()
```

```kotlin
val data = songs.filter { streams.toString().toLongOrNull() != null }
    .replace { streams } .with { streams.convertToLong() }
    .moveToLeft { streams }
    .sortByDesc { streams }
data
```

```kotlin
data.describe()
```

```kotlin
val popularArtists = data
    .groupBy("artist")
    .count()
    .sortByDesc { it["count"] }

popularArtists
```

```kotlin
popularArtists.head(20).plot { 
    bars {
        x(artist)
        y(count)
        fillColor(count){
            scale = categoricalColorHue()
        }
    }
}
```

```kotlin
data.sortByDesc { streams }.head(500).filter { artist.contains("Taylor Swift") }.count()
```

```kotlin
data.sortByDesc { streams }.head(20)
    .merge { track_name and artist }.by("-").into("Song")
    .plot {
    bars {
        y(streams)
        x("Song")
        fillColor(streams) {
            scale = categorical(listOf(Color.BLUE, Color.RED, Color.YELLOW, Color.BLACK))
        }
    }
}
```

```kotlin

val artistStats = data
    .groupBy("artist")
    .aggregate {
        count() into "count"
        mean { `danceability_%` } into "Average Danceability"
    }
    .sortByDesc { it["count"] }

artistStats
```

```kotlin
val artistStats = data
    .groupBy("mode")
    .aggregate {
        count() into "count"
        mean { `danceability_%` } into "Average Danceability"
    }
    .sortByDesc("count")

artistStats
```

```kotlin
val artistStats = data
    .groupBy { key }
    .aggregate {
        count() into "count"
    }
    .sortByDesc("count")

artistStats
```

```kotlin
val bpmData = data
    .groupBy { bpm }
    .aggregate {
        count() into "count"    
    }
    .sortByDesc { it["count"] }
    
bpmData
```

```kotlin
bpmData.filter { count > 20 }.plot {
    bars {
        y(count)
        x(bpm)
    }
}
```
