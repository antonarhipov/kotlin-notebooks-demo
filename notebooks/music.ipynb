```kotlin
%use dataframe
```

```kotlin
val music = DataFrame.readJson("../data/Spotify Extended Streaming History/Streaming_History_Audio_2023-2024_2.json")
music
```

```kotlin
val top = music
    .filter { reason_end == "trackdone" }
    .filter { master_metadata_album_artist_name !in listOf("PSY", "Little Big") }
    .groupBy { master_metadata_album_artist_name }
    .count()
    .sortByDesc("count")
top
```

```kotlin
%use kandy
```

```kotlin
top.head(10).plot {
    
    bars {
        x(master_metadata_album_artist_name)
        y(count)
        
    }
}       
```
