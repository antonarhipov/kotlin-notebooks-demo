```kotlin
val userIdA by columnOf(100, 200, 300, 400, 500)
val dataA by columnOf(10045.0, Double.NaN, 10065.0, Double.NaN, 10085.0)

val df1 = dataFrameOf("Id", "Data")(userIdA, dataA)
df1
```

```kotlin
val userIdB by columnOf(100, 200, 300, 400, 500)
val dataB by columnOf(1045.0, 130.0, 100.0, 10.0, 85.0)

val df2 = dataFrameOf("Id", "Data")(userIdB, dataB)
df2
```

```kotlin
df1.add(df2
    .rename{Id}.into("UserId")
    .rename{Data}.into("SomeData"))
```

```kotlin
df1.concat(df2)
```

```kotlin
val joined = df1.join(df2) { Id match right.Id }
joined
```
