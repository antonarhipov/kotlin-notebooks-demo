```kotlin
val amsWeather = DataFrame.readCSV("../data/ams_weather.csv", delimiter = ',')
amsWeather
```

```kotlin
amsWeather.sortBy { temp.desc() }.take(30)
```

```kotlin
amsWeather.sortBy { temp.desc() }.take(30)
    .distinct { expr { datetime.year } }
```

```kotlin
val lastYear = amsWeather.filter { datetime.year == 2023 }
    .select { datetime and nameContains("wind") }
lastYear
```
